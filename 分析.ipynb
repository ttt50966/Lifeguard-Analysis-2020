```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import locale
mpl.rc("font", family = "Noto Sans CJK TC")
```

```python
group_name = [
    "慧救",
    "水協",
    "水域",
    "水運",
    "泳救",
    "海爆",
    "海總",
    "海龍",
    "紅會",
    "輔英"
]
```

```python
總數 = []
for name in group_name:
    data = pd.read_csv(name+"_游泳池.csv")
    總數.append(len(data))
```

```python
plt.figure(dpi=300)
plt.bar(group_name,總數)
plt.title("目前體育署救生員證合格總人數")
plt.xlabel("團體")
plt.ylabel("人數 (人)")
#plt.legend(loc=(1.01,0.43))
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
```

```python
years = ["%d" %(i) for i in range(2013,2020)]
團體歷年 = []
for name in group_name:
    data = pd.read_csv(name+"_游泳池.csv")
    g = [len(data[data.證照生效日.str[:4] ==  year]) for year in years]
    團體歷年.append(g)
```

```python
plt.figure(dpi = 300)
for i in range(len(團體歷年)):
    plt.plot(years,團體歷年[i], label=group_name[i]) 
plt.xlabel("年度")
plt.ylabel("人數 (人)")
plt.title("歷年人數變化")
plt.legend(loc=(1.01,0.2))
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
```

```python
每年統計  = np.array([0])**7
for i in range(len(group_name)):
    data = pd.read_csv(group_name[i]+"_游泳池.csv")
    每年統計 = 每年統計 + np.array([len(data[data.證照生效日.str[:4] ==  year]) for year in years])
```

```python
plt.figure(dpi = 300)
for i in range(len(團體歷年)):
    plt.plot(years,每年統計) 
plt.xlabel("年度")
plt.ylabel("總人數 (人)")
plt.title("年度通過總人數")
plt.tight_layout(pad=0.4, w_pad=1.2, h_pad=1.0)
```
