```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import locale
mpl.rc("font", family = "Noto Sans CJK TC")
```

```python
group_name = [
    "慧救",
    "水協",
    "水域",
    "水運",
    "泳救",
    "海爆",
    "海總",
    "海龍",
    "紅會",
    "輔英"
]
```

```python
總數 = []
for name in group_name:
    data = pd.read_csv(name+"_游泳池.csv")
    總數.append(len(data))
```

```python
fig, ax=plt.subplots(dpi =300)
plt.tight_layout(pad=3, w_pad=4.8, h_pad=3.6)
plt.bar(group_name,總數)
plt.tick_params(direction = "in")
plt.xticks(fontsize = 12,fontweight ='regular')
plt.yticks(fontsize = 12,fontweight ='regular')
plt.setp(ax.spines.values(), linewidth =2)
plt.title("截至2020年體育署救生員合格總人數", fontsize = 14,fontweight = 'regular')
plt.xlabel("團體",fontsize = 14,fontweight ='regular')
plt.ylabel("人數 (人)",fontsize = 14,fontweight ='regular')
#plt.legend(loc=(1.01,0.43))
plt.savefig("2020年體育署游泳池救生員合格總人數.png",dpi=300)
```

```python
years = ["%d" %(i) for i in range(2013,2020)]
團體歷年 = []
for name in group_name:
    data = pd.read_csv(name+"_游泳池.csv")
    g = [len(data[data.證照生效日.str[:4] ==  year]) for year in years]
    團體歷年.append(g)
```

```python
fig, ax=plt.subplots(dpi =300)
plt.tight_layout(pad=3, w_pad=4.8, h_pad=3.6,rect=[0,0,0.94,1])
plt.tick_params(direction = "in")
plt.xticks(fontsize = 12,fontweight ='regular')
plt.yticks(fontsize = 12,fontweight ='regular')
plt.setp(ax.spines.values(), linewidth =2)
for i in range(len(團體歷年)):
    plt.plot(years,團體歷年[i], label=group_name[i]) 
plt.xlabel("年度",fontweight='regular',fontsize=12)
plt.ylabel("人數 (人)",fontweight='regular',fontsize=12)
plt.title("歷年人數變化",fontweight='regular',fontsize=14)
plt.legend(loc=(1.01,0.1),prop= {'weight':'regular'})
plt.savefig("歷年人數變化.png",dpi=300)
```

```python
每年統計  = np.array([0])**7
for i in range(len(group_name)):
    data = pd.read_csv(group_name[i]+"_游泳池.csv")
    每年統計 = 每年統計 + np.array([len(data[data.證照生效日.str[:4] ==  year]) for year in years])
```

```python
fig, ax=plt.subplots(dpi =300)
plt.tight_layout(pad=3, w_pad=4.8, h_pad=3.6)
plt.tick_params(direction = "in")
plt.xticks(fontsize = 12,fontweight ='regular')
plt.yticks(fontsize = 12,fontweight ='regular')
plt.setp(ax.spines.values(), linewidth =2)
for i in range(len(團體歷年)):
    plt.plot(years,每年統計) 
plt.xlabel("年度",fontweight ='regular',fontsize=12)
plt.ylabel("總人數 (人)",fontweight='regular',fontsize=12)
plt.title("每年通過總人數",fontsize=14,fontweight='regular')
plt.savefig("每年通過總人數.png",dpi=300)
```
